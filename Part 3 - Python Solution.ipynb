**1. Clean the names of columns to lowercase separated by “_”, remove any empty column if necessary.**

```python
# remove column Unnamed: 2
df.drop('Unnamed: 2', axis=1, inplace=True)
```

```python
# rename columns
df.columns = ['cust_id', 'join_date', 'mobiles', 'full_name']
```

**2. Change the date column to the same format ‘YYYY-MM-DD’.**

```python
df.info()
```

```python
df['join_date'] = pd.to_datetime(df.join_date.str[:10])
```

```python
df.head()
```

**3. Change the name column to the title case (e.g: Jason Mraz).**

```python
df['full_name'] = df.full_name.str.title()
```

**4. Make a new “email” column with the form:
{last_name}.{first_name}.{id}@yourcompany.com**

```python
df.head()
```

```python
df[['first_name', 'last_name']] = df.full_name.str.split(' ', expand=True)
```

```python
df.head()
```

```python
df['email'] = df.first_name + '.' + df.last_name + '.' + df.cust_id.astype('str') + '@yourcompany.com'
df.drop(['first_name', 'last_name'], axis=1, inplace=True)
```

```python
df.head()
```

**5. Change the phone number column to the format “84......”**

```python
df.mobiles
```

```python
df['mobiles'] = df.mobiles.astype('str').apply(lambda x: x if x[:2] == '84' else '84' + x)
```

```python
df.head()
```

**6. Find any duplicated ID and remove those who join later.**

```python
# check số lượng cust_id riêng biệt
df.cust_id.nunique()
```

```python
# sắp xếp dựa theo cột join_date tăng dần
df.sort_values('join_date', ascending=True, inplace=True)
```

```python
df.drop_duplicates(subset=['cust_id'], keep='first', inplace=True)
```
